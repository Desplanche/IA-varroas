## 300PPP

```python
import os
from PIL import Image

def resize_images(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Ajouter d'autres extensions d'image si nécessaire
            image_path = os.path.join(folder_path, filename)
            image = Image.open(image_path)
            image = image.resize((image.width, image.height), Image.ANTIALIAS)  # Redimensionner l'image à sa taille actuelle pour réinitialiser la résolution
            image.save(image_path, dpi=(300, 300))  # Définir la résolution à 300ppp et enregistrer l'image

# Exemple d'utilisation
folder_path = "/chemin/vers/votre/dossier"
resize_images(folder_path)
```

## RENAME

```python
import os
import glob

# Chemin du dossier contenant les fichiers
dossier = 'base_wth_gans/varroas_gans'

# Extension de sortie des fichiers renommés
extension_sortie = '.jpg'

# Liste des fichiers dans le dossier
fichiers = glob.glob(os.path.join(dossier, '*'))

# Compteur pour les noms
compteur = 1

# Parcourir chaque fichier dans la liste
for fichier in fichiers:
    # Vérifier si c'est un fichier
    if os.path.isfile(fichier):
        # Obtenir le chemin complet du fichier
        chemin_fichier = os.path.abspath(fichier)
        
        # Créer le nouveau nom de fichier
        nouveau_nom = f"G{compteur}{extension_sortie}"
        
        # Obtenir le chemin complet du nouveau fichier
        nouveau_chemin = os.path.join(dossier, nouveau_nom)
        
        # Renommer le fichier
        os.rename(chemin_fichier, nouveau_chemin)
        
        # Incrémenter le compteur
        compteur += 1
```

## RESIZE

```python
from PIL import Image
import os



def resize_images(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Ajouter d'autres extensions si nécessaire
            path = os.path.join(directory, filename)
            image = Image.open(path)
            resized_image = image.resize((22, 22))
            resized_image.save(path)

# Utilisation de la fonction
directory_path = "base_wth_gans/non_varroas/"
resize_images(directory_path)
```

## AJOUT DE BLANC POUR RENDRE LES IMAGES EN 40*40

```python
from PIL import Image
import os

def resize_images_with_white(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Ajouter d'autres extensions si nécessaire
            path = os.path.join(directory, filename)
            image = Image.open(path)

            # Création d'une nouvelle image de taille 40x40 avec fond blanc
            new_image = Image.new("RGB", (40, 40), (255, 255, 255))

            # Calcul des coordonnées pour centrer l'image de 22x22 dans la nouvelle image
            x = (40 - 22) // 2
            y = (40 - 22) // 2

            # Collage de l'image de 22x22 dans la nouvelle image
            new_image.paste(image, (x, y))

            # Enregistrement de la nouvelle image
            new_image.save(path)

# Utilisation de la fonction
directory_path = "base_wth_gans/non_varroas"
resize_images_with_white(directory_path)
```
